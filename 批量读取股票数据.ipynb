```python
import tushare as ts

# 假设您有一组股票代码
stock_codes = ["002625", "000637", "600019", "600028", "600029"]

# 为每个股票代码获取数据
stock_data = {}
for code in stock_codes:
    df = ts.get_k_data(code, start="2016-01-01", end="2023-05-22", ktype="D")
    stock_data[code] = df

# 此时，stock_data是一个字典，其中的每个键是一个股票代码，对应的值是一个包含该股票数据的DataFrame
```

```python
stock_data
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 绘制每个股票的收盘价
for code, df in stock_data.items():
    df['date'] = pd.to_datetime(df['date'])  # 转换日期列为日期格式
    plt.figure(figsize=(14, 7))
    plt.plot(df['date'], df['close'])
    plt.title(f'Stock {code} Close Price')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.grid(True)
    plt.show()
```
